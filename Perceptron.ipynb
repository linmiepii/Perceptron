```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
iris = sns.load_dataset('iris')
iris = iris[iris['species'] != 'virginica']
iris.head()
iris.drop(['sepal_width', 'petal_width'], axis = 1, inplace = True)
species_target = {
    'setosa': 1., 
    'versicolor': -1.
}
iris['target'] = iris['species'].map(species_target)
```

```python
def sign(z):
    return 1 if z>0 else -1
```

```python
# w0*x0 + w1*x1 + w2*x2 = 0
w = np.array([0., 0., 0.])
# w = [0., 0., 0.]
iterator = 0
error = 1

while error != 0:
    error = 0
    for i in range(len(iris)):
        x = np.concatenate((np.array([1.]), np.array(iris.iloc[i][:2])))
        y = np.array(iris.iloc[i][3])
        if sign(np.dot(w, x)) != y:
            iterator += 1
            error += 1     
            w = w+ y*x
            vx = np.linspace(0, 7, 20)
            vy = -(w[1] / w[2]) * vx - (w[0] / w[2])
            print('iteration = {}'.format(iterator))
            print('w = [{}]'.format(w))
            plt.plot(vx, vy, color = 'red')
            sns.scatterplot(x = 'sepal_length', y = 'petal_length', data = iris, hue = 'species')
            plt.show()
            
```
